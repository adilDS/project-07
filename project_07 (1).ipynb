```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit?

correlation = ds['Size (cm)'].corr(ds['Weight (g)'])

# Print the correlation coefficient
print(f"The correlation between Size (cm) and Weight (g) is: {correlation}")
```

```python
#visual represevtation
plt.scatter(ds['Size (cm)'], ds['Weight (g)'])
plt.xlabel('Size (cm)')
plt.ylabel('Weight (g)')
plt.title('Correlation between Size (cm) and Weight (g)')
plt.show()
```

```python
# How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))?
correlation = ds['Softness (1-5)'].corr(ds['Ripeness (1-5)'])

# Print the correlation coefficient
print(f"The correlation between Softness (1-5) and Ripeness (1-5) is: {correlation}")
```

```python
#for visual representation

plt.scatter(ds['Softness (1-5)'], ds['Ripeness (1-5)'], alpha=0.7)

plt.xlabel('Softness (1-5)')
plt.ylabel('Ripeness (1-5)')
plt.title('Relationship between Softness and Ripeness')

plt.show()
```

```python
# Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels?

data = ds.groupby('Blemishes (Y/N)').agg({'Ripeness (1-5)': 'mean', 'Softness (1-5)': 'mean'})

data
```
